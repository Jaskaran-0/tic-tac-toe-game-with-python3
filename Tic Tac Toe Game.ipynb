```python
row1=[" ", " ", " "]
row2=[" ", " ", " "]
row3=[" ", " ", " "]
```

```python
def tic_tac_toe():
    
    gamewon=False
    turn=1

    rownumber=-1
    indexnumber=-1
    char="X"
    print("Player1 has character \"X\" ")
    print("Player2 has character \"O\" ")
    
    while gamewon==False:
        
        # checking the player's turn, if odd then turn is of 1st player, if even then of 2nd player
        if turn%2!=0:
            char='X'
        else:
            char="O"
            
        # getting the row number(ie the desired list) and the index number of the list from the user     
        rownumber=int(input("Please enter the row number-> "))
        indexnumber=int(input("Please enter the index number-> "))
        
        # displaying the character inside the desired list
        
        # check if the row number is valid, ie between 1 to 3, 
        # if not between 1 and three, move to the next iteration and ask for the rownumber again
        if rownumber<1 or rownumber>3:
            print("Out of Bounds")
            continue
        else:
            #the player
             # check if the index number inside the desired list is valid, ie between 0 to 2, 
            # if not between 0 and 2, move to the next iteration and ask for the index number again
            if indexnumber<0 or indexnumber>2:
                print("Out of Bounds")
                continue
            else:
                # now that the index number and the row number are valid make changes
                # if the row number was one, insert the character in the respective row list only if the if it were blank " "
                if rownumber==1:
                    if row1[indexnumber]==" ":
                        row1[indexnumber]=char
                        # jump to the next turn
                        turn+=1
                    else:
                        print("Can't overwrite the char!")
                        continue
                    
                elif rownumber==2:
                    if row2[indexnumber]==" ":
                        row2[indexnumber]=char
                        turn+=1
                    else:
                        print("Can't overwrite the char!")
                        continue
                elif rownumber==3:
                    if row3[indexnumber]==" ":
                        row3[indexnumber]=char
                        turn+=1
                    else:
                        print("Can't overwrite the char!")
                        continue
                        
        

                
            
                
        print(row1)
        print(row2)
        print(row3)
        print(f"Turn-> {turn-1}") 
        print(" ")
        print(" ")
        
        if turn>9:
            print("It is a tie")
            break
        else:
            # check if someone is a winner or not
            # check row1
            if row1==["X", "X", "X"]:
                print("Player1 has won")
                gamewon=True
            elif row1==["O", "O", "O"]:
                print("Player2 has won")
                gamewon=True

            #check row2
            elif row2==["X", "X", "X"]:
                print("Player1 has won")
                gamewon=True
            elif row2==["O", "O", "O"]:
                print("Player2 has won")
                gamewon=True

            #check row3
            elif row3==["X", "X", "X"]:
                print("Player1 has won")
                gamewon=True
            elif row3==["O", "O", "O"]:
                print("Player2 has won")
                gamewon=True
            
            #checking diagonally from left to right
            elif row1[0]==row2[1]==row3[2]:
                if row1[0]=='X':
                    print("Player1 has won")
                    gamewon=True
                elif row1[0]=='O':
                    print("Player2 has won")
                    gamewon=True
                else:
                    continue

            
            #checking diagonally from right to left
            elif row1[2]==row2[1]==row3[0]:
                if row1[2]=='X':
                    print("Player1 has won")
                    gamewon=True
                elif row1[2]=='O':
                    print("Player2 has won")
                    gamewon=True
                else:
                    continue

                
            #checking diagonally from top to bottom for column 1
            elif row1[0]==row2[0]==row3[0]:
                if row1[0]=='X':
                    print("Player1 has won")
                    gamewon=True
                elif row1[0]=='O':
                    print("Player2 has won")
                    gamewon=True
                else:
                    continue

            
            #checking diagonally from top to bottom for column 2
            elif row1[1]==row2[1]==row3[1]:
                if row1[1]=='X':
                    print("Player1 has won")
                    gamewon=True
                elif row1[1]=="O":
                    print("Player2 has won")
                    gamewon=True
                else:
                    continue

            
            #checking diagonally from top to bottom for column 3
            elif row1[2]==row2[2]==row3[2]!=" ":
                if row1[2]=='X':
                    print("Player1 has won")
                    gamewon=True
                elif row1[2]=='O':
                    print("Player2 has won")
                    gamewon=True
                else:
                    continue
            else:
                continue
            
                
                    
```
